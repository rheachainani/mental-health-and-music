```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

client_id = '5f5a4386049947f6ab225000e959df25'
client_secret = 'b3d6e5f5379643f2949922c2be175690'
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

playlist_id = 'spotify:playlist:4muHyvSwG1wP9sI4XihC5w'

results = sp.playlist_tracks(playlist_id)


track_names = []
cover_arts = []
artist_name = []
genres = []
for item in results['items']:
    if item['track'] is not None and item['track']['type'] == 'track':
        track = item['track']
        track_names.append(track['name'])
        cover_arts.append(track['album']['images'][0]['url'])
        artist_name.append(track['artists'][0]['name'])
        artist_info = sp.artist(track['artists'][0]['id'])
        genres.append(', '.join(artist_info['genres'])) if artist_info['genres'] else genres.append('N/A')

data = {
    'Artist Name': artist_name,
    'Track Name': track_names,
    'Cover Art URL': cover_arts,
    'Genre': genres
}
df = pd.DataFrame(data)
```
